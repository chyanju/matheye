```python
from pathlib import Path
from natsort import natsorted

root_folder = Path("../benchmarks/minif2f-decomposed-ngram/")
ngram_files = natsorted(list(root_folder.rglob("*.txt")))
```

```python
import re

def clean_token(token):
    return re.sub(r'(_@?\d+)+$', '', token)
```

```python
seqs = []
for file_path in ngram_files:
    print(file_path)
    with open(file_path, "r") as f:
        raw_lines = [p.strip() for p in f.readlines()]

    for rl in raw_lines:
        # clean the postfixes
        tokens = tuple([clean_token(p) for p in rl.split(" ") if p != "have"])
        seqs.append(tokens)

# filter out empty ones
seqs = [p for p in seqs if len(p) > 0]
seqs = [p for p in seqs if p != ("",)]
```

```python
from collections import Counter
cc = Counter(seqs)

for p in cc.most_common():
    ps = ", ".join(p[0])
    print(f"{len(p[0])}, {p[1]}, {ps}")
```
