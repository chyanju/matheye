```python
import json
from pathlib import Path

# suite = "proverbench@1"
# folder = "proverbench"
suite = "minif2f@1"
folder = "minif2f"
# suite = "putnam@1"
# folder = "putnam"

result_path = f"../../lean-pl-fix/decomposition_results/{folder}_minimal_verification_summary.json"
with open(Path(result_path).expanduser(), "r") as f:
    result_json = json.load(f)
result_json = result_json["results"]

from natsort import natsorted
result_json = natsorted(result_json, key=lambda x: x["problem_id"])
```

## stepwise data

```python
skipped = []

for bigbench in result_json:

    if not bigbench["hole_verification_pass"]:
        skipped.append(bigbench["problem_id"])
    
    r = bigbench["tactic_mapping"] # unigram's result
    d = bigbench["original_tactics_test"] # dspv2's result
    for hole in natsorted(r.keys()):
        bline = [ suite, bigbench["problem_id"] ]
        bline.append(hole)
        if hole in d:
            bline.append( 1 if d[hole]["success"] else 0 )
        else:
            bline.append( -1 )
        if r[hole] == "admit":
            bline.append(0)
        else:
            bline.append(1)
        bline.append(r[hole])

        print(", ".join([f"{p}" for p in bline]))
```

## full benchmark data

```python
for bigbench in result_json:
    bline = []
    bline.append(suite)
    bline.append(bigbench["problem_id"])
    bline.append(bigbench["num_steps"] if "num_steps" in bigbench.keys() else -1)
    bline.append(1 if bigbench["original_verification_pass"] else 0)
    if "detailed_steps" in bigbench.keys():
        hps = [p["hole_verification_pass"] for p in bigbench["detailed_steps"]]
        if False in hps:
            # there's invalid sketch somewhere in the steps
            bline.append(-1)
        else:
            bline.append(1 if bigbench["synthesized_verification_pass"] else 0)
    else:
        # there's full benchmark level bug
        bline.append(-2)
    
    print(", ".join([f"{p}" for p in bline]))
    
```
